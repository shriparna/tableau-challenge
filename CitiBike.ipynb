```python
import pandas as pd
```

```python
nov_df = pd.read_csv("JC-202211-citibike-tripdata.csv")
nov_df.dropna(how='any', inplace=True)
nov_df.info()
nov_df.head()
```

```python
dec_df = pd.read_csv("JC-202212-citibike-tripdata.csv")
dec_df.dropna(how='any', inplace=True)
dec_df.info()
dec_df.head()
```

```python
jan_df = pd.read_csv("JC-202301-citibike-tripdata.csv")
jan_df.dropna(how='any', inplace=True)
jan_df.info()
jan_df.head()
```

```python
feb_df = pd.read_csv("JC-202302-citibike-tripdata.csv")
feb_df.dropna(how='any', inplace=True)
feb_df.info()
feb_df.head(10)
```

```python
mar_df = pd.read_csv("JC-202303-citibike-tripdata.csv")
mar_df.dropna(how='any', inplace=True)
mar_df.info()
mar_df.head()
```

```python
apr_df = pd.read_csv("JC-202304-citibike-tripdata.csv")
apr_df.dropna(how='any', inplace=True)
apr_df.info()
apr_df.head()
```

```python
citi_bike_df = nov_df.append(dec_df).append(jan_df).append(feb_df).append(mar_df).append(apr_df)
citi_bike_df.info()
citi_bike_df.head()
```

```python
citi_bike_df.to_csv("JC-Citi-Bike.csv")
```
